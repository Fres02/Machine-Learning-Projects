## Under Sampling

Since the number of fraud transactions is much lower than legit transactions. 
The main purpose of building this ML model is to detect fraud transactions. 
Without appropriate handling of the data imbalance, this model risks becoming biased toward predicting transactions as legit, potentially overlooking the rare instances of fraud.


```python
#Randomly selecting 492 data rows from legit. 
legit_sample = legit.sample(n=492)
```

```python
#Concatanating two data frames

new_df = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df["Class"].value_counts()
```

```python
new_df.groupby("Class").mean()
```

## Split dataset into features and target

```python
X = new_df.drop(columns="Class",axis = 1)
Y = new_df["Class"]
```

```python
X
```

```python
Y
```

# Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify = Y, random_state=2)
```

```python
X.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

# Model Training

## Logitic Regression

```python
model = LogisticRegression()
```

### Training the model with training data

```python
model.fit(X_train,Y_train)
```

# Model Evaluation

## Accuracy Score

### Accuracy on training data 

```python
X_train_prediction = model.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
training_data_accuracy
```

### Accuracy on testing data 

```python
X_test_prediction = model.predict(X_test)
```

```python
testing_test_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
testing_test_accuracy
```

The accuracy scores of training data and testing data are nearly equal. Therefore it can be assumed that model is generalized.
